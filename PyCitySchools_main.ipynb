```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Create a dataframe of school type for last groupby district/charter cell
school_type_df = school_data_complete[["school_name","type"]]
```

```python
#Calculate values for District summary dataframe

#Count schools
totalSchools = len(school_data_complete["School ID"].unique())
#Count students
totalStudents = len(school_data_complete["Student ID"].unique())
#Sum of each school's budget after unique function creates a series
totalBudget = sum(school_data_complete['budget'].unique())

#Use mean() function to average scores
averageMath = school_data_complete['math_score'].mean()
averageReading = school_data_complete['reading_score'].mean()

#Students at or aboe 70 passed into new data frame to calculate passing percentage
passingMath_df = school_data_complete.loc[school_data_complete["math_score"] >= 70]
passingMath = len(passingMath_df)
passingMath = (passingMath / totalStudents) * 100

#Students at or above 70 passed into data frame to calculate passing percentage
passingReading_df = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
passingReading = len(passingReading_df)
passingReading = (passingReading / totalStudents) * 100

#Students at or above 70 in both subjects passed into data frame to calculate passing percentage
passingOverall_df = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
passingOverall = len(passingOverall_df)
passingOverall = (passingOverall / totalStudents) * 100

# Place all of the data found into a summary DataFrame
schools_summary_df = pd.DataFrame({"Total Schools": [totalSchools],
                              "Total Students": totalStudents,
                              "Total Budget": totalBudget,
                              "Average Math Score": averageMath,
                              "Average Reading Score": averageReading,
                              "% Passing Math": passingMath,
                              "% Passing Reading":passingReading,
                              "% Overal Passing": passingOverall})
                              
#Optional formatting
schools_summary_df["Total Students"] = schools_summary_df["Total Students"].map("{:,}".format)
schools_summary_df["Total Budget"] = schools_summary_df["Total Budget"].astype(float).map("${:,.2f}".format)

schools_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Summary

#Create data frame grouped by school
byschool_df = school_data_complete
byschool_df = byschool_df.groupby(["school_name"])

#Count students for summary
studentCount = byschool_df["Student ID"].count()
#Capture budget by school (all grouped entries will have same amount)
schoolBudget = byschool_df["budget"].mean()
#Capture the type of school (all grouped entries will have same value)
schoolType = byschool_df["type"].unique()
#Divide budget by students to capture budget allocated per student
budgetPerStudent = schoolBudget / studentCount

#Use mean function to average all the math scores for that school
averageMath = byschool_df["math_score"].mean()
#Use mean function to average all the reading scores for that school
averageReading = byschool_df["reading_score"].mean()

#Using previous dataframe only containing students passing math
#Counting number of each school's name provides the number of students passing
passingM_df = passingMath_df["school_name"]
passingM = passingM_df.value_counts()
#Show as percentage
passingM = passingM / studentCount * 100

#Using dataframe from previous cell only containing students passing reading
#Counting number of each school's name provides the number of students passing
passingR_df = passingReading_df["school_name"]
passingR = passingR_df.value_counts()
#Show as percentage
passingR = passingR / studentCount * 100

#Using dataframe from previous cell only containing students passing overall
#Counting number of each school's name provides the number of students passing
passingO_df = passingOverall_df["school_name"]
passingO = passingO_df.value_counts()
#Show as percentage
passingO = passingO / studentCount * 100

byschool_summary_df = pd.DataFrame({"School Type": schoolType,
                                    "Total Students": studentCount,
                                    "School Budget": schoolBudget,
                                    "Per Student Budget": budgetPerStudent,
                                    "Average Math Score": averageMath,
                                    "Average Reading Score": averageReading,
                                    "% Passing Math": passingM,
                                    "% Passing Reading": passingR,
                                    "% Overall Passing": passingO})

#Optional formatting
byschool_summary_format_df = byschool_summary_df
byschool_summary_format_df["School Budget"] = byschool_summary_format_df["School Budget"].astype(float).map("${:,.2f}".format)
#Removed formating on per Student budget so dataframe could be used later on for binning
#byschool_summary_format_df["Per Student Budget"] = byschool_summary_format_df["Per Student Budget"].astype(float).map("${:,.2f}".format)

byschool_summary_format_df

```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create panda series for each grade
ninth_df = school_data_complete.loc[school_data_complete["grade"] == "9th"]
ninth_df = ninth_df.groupby(["school_name"])
ninthMAverage = ninth_df["math_score"].mean()

tenth_df = school_data_complete.loc[school_data_complete["grade"] == "10th"]
tenth_df = tenth_df.groupby(["school_name"])
tenthMAverage = tenth_df["math_score"].mean()

eleventh_df = school_data_complete.loc[school_data_complete["grade"] == "11th"]
eleventh_df = eleventh_df.groupby(["school_name"])
eleventhMAverage = eleventh_df["math_score"].mean()

twelfth_df = school_data_complete.loc[school_data_complete["grade"] == "12th"]
twelfth_df = twelfth_df.groupby(["school_name"])
twelfthMAverage = twelfth_df["math_score"].mean()

#Create new summary data frame
math_bygrade_df = pd.DataFrame({"9th": ninthMAverage,
                                    "10th": tenthMAverage,
                                    "11th": eleventhMAverage,
                                    "12th": twelfthMAverage})

math_bygrade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create the bins in which Data will be held
# Bins are (0, 585),(586, 630), (631, 645), (646,680).   
spending_bins = [0, 585, 630, 645, 680]

# Create the names for the five bins
spending_group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

#Use data frame from school summary cell for budget breakdown
bybudget_summary_df = byschool_summary_df

#Set bins
bybudget_summary_df["Spending Ranges(Per Student)"] = pd.cut(bybudget_summary_df["Per Student Budget"], spending_bins, labels=spending_group_names, include_lowest=True)
#Group by new bin column
bybudget_summary_df = bybudget_summary_df.groupby(["Spending Ranges(Per Student)"])

#Use mean function to calculate averages required columns
bybudget_summary_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
```

##Observations

* It's a common conception that the lack of overall student success at schools is directly linked to limited budgets that don't allow for the sufficient resources for students.  The trend in the output above directly contradicts that conception.  Regardless of the school type, the more that was spent on a student per capita the lower the overall performances of the students in terms of average scores and passing rates. 

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Create the bins in which Data will be held
# Bins are (0, 999),(1000, 1999), (2000, 5000).   
school_size_bins = [0, 999, 1999, 5000]

# Create the names for the five bins
school_size_names = ["Small(<1000)", "Medium(1000-1999)", "Large(2000-5000)"]

#Use data frame from school summary cell for school size breakdown
bysize_summary_df = byschool_summary_df
bysize_summary_df["School Size"] = pd.cut(bysize_summary_df["Total Students"], school_size_bins, labels=school_size_names, include_lowest=True)
#Group by new school size column
bysize_summary_df = bysize_summary_df.groupby(["School Size"])

#Use mean function to calculate averages required columns
bysize_summary_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
```
